```python
import datetime
import glob
import pandas as pd
```

```python
def load_data(date=None, date_from=None, date_to=None):

    data_list = glob.glob('../rawdata/*-tripdata.csv.zip')
    data_list.sort()
    def read_csv(filename):
        return pd.read_csv(filename, compression='zip')
    
    if date is not None:
        filename = '../rawdata/{}-tripdata.csv.zip'.format(date)
        return read_csv(filename)
        
    if date is None and date_from is None and date_to is None:
        return read_csv(data_list[0])
```
